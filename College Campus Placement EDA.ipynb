## Counting Values

```python
data['name of company'].value_counts()
```

```python
data['college name'].value_counts()
```

```python
data['region'].value_counts()
```

```python
mum=data[data['region']=='MUMBAI']['name of company'].count()
nmum=data[data['region']=='NAVI MUMBAI']['name of company'].count()
pun=data[data['region']=='PUNE']['name of company'].count()
tha=data[data['region']=='THANE']['name of company'].count()
region=[['Mumbai',mum],['Navi Mumbai',nmum], ['Pune',pun],['Thane',tha]]
region=pd.DataFrame(region,columns=['region','count'])
region
```

## Plotting Bar of Region vs Count

```python
sns.barplot(x='region',y='count',data=region)
```

## Extracting Mumbai Region Data From Dataset

```python
df = data[data['region']=='MUMBAI'].sort_values('name of company')
df.head()
```

```python
df.shape
```

```python
df['name of company'].value_counts()
```

```python
df['college name'].value_counts()
```

## Plotting Bar Plot College vs Salary

```python
fig = plt.figure(figsize=(10, 5))
  
plt.bar('college name', 'Salary', color='maroon', width=0.4, data=df)
  
plt.xlabel("College Names")
plt.xticks(rotation=90)
plt.ylabel("Salary")
plt.show()
```

## Plotting Bar Plot Company vs Salary

```python
fig = plt.figure(figsize=(35, 5))
  
plt.bar('name of company', 'Salary', color='maroon', width=0.4, data=df)
  
plt.xlabel("Company")
plt.xticks(rotation=90)
plt.ylabel("Salary")
plt.show()
```

## Plotting Bar Plot Company vs College

```python
fig = plt.figure(figsize=(35, 5))
  
plt.bar('name of company', 'college name', color='maroon', width=0.4, data=df)
  
plt.xlabel("Company")
plt.xticks(rotation=90)
plt.ylabel("Salary")
plt.show()
```
